```python
from langchain.document_loaders import PyPDFLoader
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
loader_docker_book = PyPDFLoader("/content/drive/MyDrive/AI_integration_kernel/Docker Deep Dive by Nigel Poulton.pdf")
```

```python
loader_podman_book = PyPDFLoader("/content/drive/MyDrive/AI_integration_kernel/Podman-in-Action-ebook-FINAL-Red-Hat-Developer.pdf")
```

```python
docker_documents = loader_docker_book.load()
```

```python
podman_documents = loader_podman_book.load()
```

```python
print(len(docker_documents))
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=100)
```

```python
text_docker = text_splitter.split_documents(docker_documents)
```

```python
text_podman = text_splitter.split_documents(podman_documents)
```

```python
print(len(text_docker))
```

```python
from langchain.embeddings import HuggingFaceEmbeddings, OpenAIEmbeddings,HuggingFaceBgeEmbeddings
```

```python
hf_embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
hf_bge_embeddings = HuggingFaceBgeEmbeddings(model_name="BAAI/bge-large-en")
```

```python
from google.colab import userdata
OPENAI_API_KEY=userdata.get('OpenAIKey')
```

```python
import os
os.environ["OPENAI_API_KEY"]=OPENAI_API_KEY
```

```python
openai_embeddings = OpenAIEmbeddings(openai_api_key=OPENAI_API_KEY)
```

```python
from langchain.vectorstores import Chroma
import chromadb
```

```python
import os
os.getcwd()
```

```python
CURRENT_DIR = os.path.dirname(os.path.abspath("."))
CURRENT_DIR
```

```python
DB_DIR = os.path.join(CURRENT_DIR, "/content/db")
```

```python
DB_DIR
```

```python
client_settings = chromadb.config.Settings(
    is_persistent=True,
    persist_directory=DB_DIR,
    anonymized_telemetry=False,
)
```

```python
harrypotter_vectorstore = Chroma.from_documents(text_docker,
                                       hf_bge_embeddings,
                                       client_settings=client_settings,
                                       collection_name="dockerBook",
                                       collection_metadata={"hnsw":"cosine"},
                                       persist_directory="/store/docker_book")
```

```python
podman_vectorstore = Chroma.from_documents(text_podman,
                                           hf_bge_embeddings,
                                           client_settings=client_settings,
                                           collection_name="podmanBook",
                                           collection_metadata={"hnsw":"cosine"},
                                           persist_directory="/store/podman_book")
```

```python
retriever_harrypotter = harrypotter_vectorstore.as_retriever(search_type="mmr",search_kwargs={"k": 5, "include_metadata": True})
```

```python
retriever_podman = podman_vectorstore.as_retriever(search_type="mmr",search_kwargs={"k": 5, "include_metadata": True})
```

```python
from langchain.retrievers.merger_retriever import MergerRetriever
```

```python
lotr = MergerRetriever(retrievers=[retriever_harrypotter, retriever_podman])
```

```python
for chunks in lotr.get_relevant_documents("What is Image?"):
    print(chunks.page_content)
```
