```python
from langchain.embeddings import HuggingFaceEmbeddings, OpenAIEmbeddings,HuggingFaceBgeEmbeddings
```

```python
hf_embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
hf_bge_embeddings = HuggingFaceBgeEmbeddings(model_name="BAAI/bge-large-en")
```

```python
from langchain.vectorstores import Chroma
import chromadb
```

```python
import os
os.getcwd()
```

```python
CURRENT_DIR = os.path.dirname(os.path.abspath("."))
CURRENT_DIR
```

```python
DB_DIR = os.path.join(CURRENT_DIR, "/content/db")
```

```python
DB_DIR
```

```python
client_settings = chromadb.config.Settings(
    is_persistent=True,
    persist_directory=DB_DIR,
    anonymized_telemetry=False,
)
```

```python
harrypotter_vectorstore = Chroma.from_documents(text_docker,
                                       hf_bge_embeddings,
                                       client_settings=client_settings,
                                       collection_name="dockerBook",
                                       collection_metadata={"hnsw":"cosine"},
                                       persist_directory="/store/docker_book")
```

```python
podman_vectorstore = Chroma.from_documents(text_podman,
                                           hf_bge_embeddings,
                                           client_settings=client_settings,
                                           collection_name="podmanBook",
                                           collection_metadata={"hnsw":"cosine"},
                                           persist_directory="/store/podman_book")
```

```python
retriever_harrypotter = harrypotter_vectorstore.as_retriever(search_type="mmr",search_kwargs={"k": 5, "include_metadata": True})
```

```python
retriever_podman = podman_vectorstore.as_retriever(search_type="mmr",search_kwargs={"k": 5, "include_metadata": True})
```

```python
from langchain.retrievers.merger_retriever import MergerRetriever
```

```python
lotr = MergerRetriever(retrievers=[retriever_harrypotter, retriever_podman])
```

```python
for chunks in lotr.get_relevant_documents("What is Image?"):
    print(chunks.page_content)
```

```python
for chunks in lotr.get_relevant_documents("What is daemon?"):
    print(chunks.page_content)
```

```python
from langchain.document_transformers import (
    EmbeddingsClusteringFilter,
    EmbeddingsRedundantFilter,
)
from langchain.retrievers.document_compressors import DocumentCompressorPipeline
from langchain.retrievers import ContextualCompressionRetriever
from langchain.document_transformers import LongContextReorder
```

```python
from re import search
filter = EmbeddingsRedundantFilter(embeddings=hf_bge_embeddings)
reordering = LongContextReorder()
pipeline = DocumentCompressorPipeline(transformers=[filter, reordering])
compression_retriever_reordered = ContextualCompressionRetriever(
    base_compressor=pipeline, base_retriever=lotr,search_kwargs={"k": 3, "include_metadata": True}
)
```

```python
docs = compression_retriever_reordered.get_relevant_documents("What is podman?")
print(len(docs))
print(docs[0].page_content)
```

```python
from langchain.llms import LlamaCpp
llms = LlamaCpp(streaming=True,
                   model_path="/content/drive/MyDrive/AI_integration_kernel/zephyr-7b-beta.Q4_0.gguf",
                   max_tokens = 1500,
                   temperature=0.75,
                   top_p=1,
                   gpu_layers=0,
                   stream=True,
                   verbose=True,n_threads = int(os.cpu_count()/2),
                   n_ctx=4096)
```

```python
from langchain.chains import RetrievalQA
```

```python
qa = RetrievalQA.from_chain_type(
      llm=llms,
      chain_type="stuff",
      retriever = compression_retriever_reordered,
      return_source_documents = True
)
```

```python
query ="tell me about docker.inc"
results = qa.invoke(query)
print(results['result'])
#
print(results["source_documents"])
```

```python
results = qa.invoke("what is podman?")
print(results['result'])
#
print(results["source_documents"])
#
for source in  results["source_documents"]:
    print(source.metadata)
```
